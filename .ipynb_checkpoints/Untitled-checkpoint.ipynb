```python
# now we have frames of videos

# we need to convert them to grayscale


import cv2
import matplotlib.pyplot as plt
import numpy as np
# reading a single image

import glob
cv_img = []
count = 0
```

```python
images = []
files = []
for img in glob.glob("frames/*.jpg"):
    file = img
    # read image
    img = cv2.imread(img)
    images.append(img)
    files.append(file)
    
    
```

```python
print(len(images))
```

```python
import re
newfiles = []
for file in files:
    newfile = re.findall(r'\d+', file)
    newfiles.append(int(newfile[0]))
    #print(file, newfile)
    
```

```python
dicc = dict()
count = 0
for file in newfiles:
    dicc[file] = count
    count += 1
```

```python
# this will make the mapping correct

from matplotlib import pyplot as plt

for i in range(2):
    img = images[dicc[i]]

    
    plt.imshow(img,cmap='gray')
    plt.title(i)
    plt.show()    
```

```python
croped_rotate = []

for img in images:
    hh, ww = img.shape[:2]



    # define circles
    radius1 = 200
    xc = int(1950/2)
    yc = int(1250/2)

    # draw filled circles in white on black background as masks
    mask1 = np.zeros_like(img)
    mask1 = cv2.circle(mask1, (xc,yc), radius1, (255,255,255), -1)
    mask = mask1

    # put mask into alpha channel of input
    result = cv2.cvtColor(img, cv2.COLOR_BGR2BGRA)
    result[:, :, 3] = mask[:,:,0]

    crop = result[yc-radius1:yc+radius1, xc-radius1:xc+radius1,]

    height, width = crop.shape[:2]
    center = (width/2, height/2)
    rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=-50, scale=1)
    rotated_image = cv2.warpAffine(src=crop, M=rotate_matrix, dsize=(width, height), borderValue=(255, 255, 255))
    
    
    # COVERTING TO GRAYSCALE
    gray = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2GRAY)

    #ret, thres = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
    croped_rotate.append(gray)

```

```python
print(len(croped_rotate))
print(croped_rotate[1].shape)
```

```python
from matplotlib import pyplot as plt

thress = []



for i in range(61, len(images)):
    img = croped_rotate[dicc[i]]
    crop_img = img[150:200, 150:250]
    ret, thres = cv2.threshold(crop_img,50,255,cv2.THRESH_BINARY)
    thress.append(thres)
    
    plt.imshow(img,cmap='gray')
    plt.title(i)
    plt.show()
    plt.imshow(thres,cmap='gray')
    plt.title(i)
    plt.show()
    

    
    
    image = thres
    img_row_sum = np.sum(255-image,axis=0).tolist()
    plt.plot(img_row_sum)
    plt.show() 
    
    
```

```python
# get all points above 8000

highpointmarker = 9000
lowpointmarker = 500

vertical_np = np.array(vertical_hist)
print(vertical_np.shape)
```

```python
results = []
bottoms = []
for i in range(vertical_np.shape[0]):
    res = np.asarray(vertical_np[i]>highpointmarker).nonzero()[0]
    bot = np.asarray(vertical_np[i]<lowpointmarker).nonzero()[0]
    results.append(res[-1] - res[0])
    
    semibot1 = np.asarray(vertical_np[:res[0]]<lowpointmarker).nonzero()[0][-1]
    semibot2 = np.asarray(vertical_np[res[-1]:]<lowpointmarker).nonzero()[0][0] + res[-1]
    print(res[0], res[-1], semibot1, semibot2)
    #print(vertical_np[res[-1]:]<lowpointmarker)
    bottoms.append(semibot2 - semibot1)
    
    

```
